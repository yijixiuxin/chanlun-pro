```python
from cl_v2 import cl
from cl_v2 import kcharts
from cl_v2 import exchange_binance
from cl_v2 import exchange_db
from prettyprinter import cpprint
import time, datetime
import pandas as pd
```

```python
code = 'BTC/USDT'
frequencys = ['d', '30m']
start_date = None
end_date = None
# exchange = exchange_binance.ExchangeBinance()
exchange = exchange_db.ExchangeDB('currency')
```

```python
renders = []
klines = {}
for f in frequencys:
    _ks = exchange.klines(code, frequency=f, start_date = start_date, end_date = end_date, args={'limit': 1000})
#     _ks = kline_ks0:988:]
    print(f, '获取K线数据量：', len(_ks))
    klines[f] = _ks

cl_datas = cl.batch_cls(code, klines)
for cd in cl_datas:
    title = '%s - 【%s】 周期数据图表' % (code, cd.frequency)
    renders.append(kcharts.render_charts(title, cd))
```
